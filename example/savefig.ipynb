```python
import numpy as np
import matplotlib.pyplot as mp

n = 5000
x = np.random.normal(0, 1, n)
y = np.random.normal(0, 1, n)

mp.figure('Gaussian')
mp.scatter(x, y, s=60, marker='.')
mp.show()
```

```python
d = x**2 + y**2
mp.figure('Gaussian2')
mp.scatter(x, y, s=30, marker='.', c=d, cmap='jet_r')
```

```python
d = x**2 + y**2
mp.figure('Gaussian')
mp.scatter(x, y, s=30, marker='.', c=d, cmap='jet_r')
mp.savefig('Gaussian.jpg', dpi=1200)
```
